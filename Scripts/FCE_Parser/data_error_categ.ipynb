```python
import numpy as np
import pandas as pd
```

```python
df_data = pd.read_csv("../data/fce_txt/all_converted.csv")
```

```python
df_data
```

```python
df_error_cat = pd.read_csv("error_categories.csv")
```

```python
df_error_cat
```

```python
error_map = dict(zip(df_error_cat["Error Code"],df_error_cat["Category Code"]))
```

```python
def map_errors(row, error_map): 
    return error_map[row['error_tag']]
```

```python
df_data['err_type'] = df_data.apply (lambda row: map_errors(row, error_map), axis=1)
```

```python
df_data
```

```python
df_data.to_csv("categorized_data.csv", header = True, index=False)
```
